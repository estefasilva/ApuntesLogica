```python
import pandas as pd
```

```python
data = {"costumer_id": [1,2,3], "costumer_name":["John", "Anna", "Peter"],"costumer_salary":[1200000, 1250000, 1300000]}
pd.DataFrame(data)
```

```python
costumer = pd.DataFrame(data)
print(costumer)
```

```python
costumer.to_csv("costumers.csv", index= False)
```

```python
costumer.to_excel("costumers.xlsx")
```

```python
costumer.to_json("costumers.json")
```
